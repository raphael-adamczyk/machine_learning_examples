```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 300
D = 2
K = 3
```

```python
X = np.zeros([N, D])
X[:100, :] = np.random.randn(100, D) + np.array([0, 0])
X[100:200, :] = np.random.randn(100, D) + np.array([5, 5])
X[200:300, :] = np.random.randn(100, D) + np.array([0, 5])
X
```

```python
Y = np.array([0]*100 + [1]*100 + [2]*100)
Y
```

```python
graph = plt.scatter(X[:, 0], X[:, 1], c=Y)
graph
```

```python
mean_x0 = sum(x for x in X[:100, :])/100
mean_x0
```

```python
mean_x1 = sum(x for x in X[100:200, :])/100
mean_x1
```

```python
mean_x2 = sum(x for x in X[200:300, :])/100
mean_x2
```

```python
V_mean = np.array([mean_x0, mean_x1, mean_x2])
V_mean
```

```python
X.mean(axis=0)
```

```python
means = np.zeros([K, D])
means[0, :] = X[Y == 0].mean(axis=0)
means[1, :] = X[Y == 1].mean(axis=0)
means[2, :] = X[Y == 2].mean(axis=0)
```

```python
means
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.scatter(means[:, 0], means[:, 1], marker='*', s=500, c="red")
```
